```python
#Question5

#The url_for function is used in Flask for URL building. It takes the name of a function as its first argument
#and any number of keyword arguments each corresponding to a variable part of the URL rule

from flask import Flask, url_for

app = Flask(__name__)

@app.route("/hello/<name>")
def hello(name):
    return "Hello, {}!".format(name)

url = url_for("hello", name="shaury")
print(url)
```
